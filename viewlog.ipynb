```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('pretrain_log_epoch_0_4.csv')
df2 = pd.read_csv('gan_training_log_epoch_5_plus.csv')
# df2 = pd.read_csv('training_data.csv')
```

```python
df1.head(5)
```

```python
plt.figure(figsize=(12, 5))

plt.plot(df1.index, df1['Pre_train_G_loss'], label="Pre_train_G_loss")

plt.xlabel("Step")
plt.ylabel("Pre_train_G_loss")
plt.title("Pre_train_G_loss")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.show() 
```
